#### Vemos el % en las columnas 'sentiment_analysis' y 'reviews_recommend'

```python
def resumen_cant_porcentaje(df, columna):
    """
    Genera un resumen de cantidad y porcentaje para una columna específica en un DataFrame.

    Parámetros:
    - df: DataFrame de pandas
    - columna: Nombre de la columna para la cual se desea el resumen

    Imprime el resumen en la consola.
    """
    # Lógica para resumen de cantidad y porcentaje
    # Ajusta esto según tus necesidades específicas
    total = len(df)
    counts = df[columna].value_counts()
    percentages = (counts / total) * 100

    # Imprimir resumen
    print(f"Resumen para columna '{columna}':")
    for valor, count, porcentaje in zip(counts.index, counts, percentages):
        print(f"{valor}: {count} ({porcentaje:.2f}%)")

# Ahora puedes utilizar esta función en tu DataFrame
resumen_cant_porcentaje(df_user_reviews, 'sentiment_analysis')
resumen_cant_porcentaje(df_user_reviews, 'reviews_recommend')
```
